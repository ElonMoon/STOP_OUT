### 12. 'url' 주소에 있는 공지사항에 해당하는 'Notice' 클래스를 만들고, 크롤링 시 공지 사항 하나마다 'Notice'클래스 인스턴스를 만들어 'notice_list'에 추가한다.

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
BASE_URL = 'http://www.leagueoflegends.co.kr/?m=news&cate=notice'
response = requests.get(BASE_URL)
```

```python
html = response.text
soup = BeautifulSoup(html, 'lxml')
```

```python
soup
```

```python
notice_table = soup.select_one('div.left-contents > div.border-box > div.contents-table')
```

```python
notice_table
```

```python
tr_list = notice_table.select_one('table > tbody').select('tr')
```

```python
for tr in tr_list:
    title = tr.select_one('td > a ').get_text(strip=True)
    print(title)
```

```python
class Notice:
    def __init__(self, title, date, view):
        self.title = title
        self.date = date
        self.view = view
```

```python
notice_list = []
for tr in tr_list:
    td = tr.select('td')
    title = td[0].select_one('a').get_text(strip=True)
    created_date = td[1].get_text(strip=True)
    view_counts = td[2].get_text(strip=True)
    new_notice = Notice(
        title,
        created_date,
        view_counts)
    notice_list.append(new_notice)
```

```python
for notice in notice_list:
    print(notice.title)
    print(notice.date)
    print(notice.view)
    print()
```
